# Section 3: Powering Decisions (Feature Extraction Needs)

## 21.	The board wants a daily total of renewable energy production.

```python
df_cleaned['daily_total'] = df_cleaned.groupby(df['timestamp'].dt.date)['energy_source_mix'].transform(sum)
```

## 22.	Executives need to compare grid dependence against renewable contribution.

```python
df_cleaned[['grid_frequency', 'energy_source_mix']].corr()
```

## 23.	Forecasting accuracy must be quantified â€” identify deviation between projected and actual demand.

```python
df_cleaned['simulated_actual_demand'] = (
    df_cleaned['temperature'] * 0.4 +
    df_cleaned['humidity'] * 0.2 +
    df_cleaned['price_signal'] * 0.3 +
    df_cleaned['voltage_level'] * 0.1
)
```

```python
df_cleaned['deviation'] = (df_cleaned['historical_avg_demand'] - df_cleaned['simulated_actual_demand']).abs()
```

## 24.	Timed marketing campaigns require labels for morning, afternoon, evening slots.

```python
df_cleaned['slot'] = df_cleaned['timestamp'].apply(lambda x: 'Morning' if 4 < x.hour <= 12 else ('Afternoon' if 12 < x.hour <= 19 else 'Evening'))
```

## 25.	Climate adaptation reports need clear segmentation of temperature zones.

```python
df_cleaned['temperature_zone'] = df_cleaned['temperature'].apply(lambda x: 'Low' if x <= 0.2 else ('Medium' if 0.20 < x <= 0.80 else 'High'))
```

```python
df_cleaned
```

## 26.	Grid reliability teams want voltage segmented into operational stability bands.

```python
df_cleaned['operational_bands'] = df_cleaned['voltage_level'].apply(lambda x: 'Operational' if 0.20 <= x <=0.80 else 'Non Operational')
```

```python
df_cleaned
```

## 27.	Humidity trends are suspected to affect performance â€” define the thresholds.

```python
df_cleaned['Affected_performance'] = df_cleaned['humidity'].apply(lambda x: 'Not Affected' if 0.20 <= x <=0.80 else 'Affected')
```

## 28.	Safety committees request flags for overload risk events.

```python
df_cleaned['overload_flag'] = df_cleaned['voltage_level'] > 0.8
```

```python
df_cleaned
```

## 29.	Finance needs insight into high-consumption windows and pricing overlap.

```python
df_cleaned[df_cleaned['simulated_actual_demand'] > 0.7][['simulated_actual_demand', 'price_signal']].corr()
```

## 30.	Demand teams want an intuitive metric for consumption efficiency.

```python
df_cleaned['Efficiency_Ratio'] = df_cleaned['historical_avg_demand'] / df_cleaned['voltage_level']
df_cleaned['Efficiency_interpretation'] = df_cleaned['Efficiency_Ratio'].apply(lambda x:'Highly Efficient' if x >= 0.5 else 'Low Efficient')
```

# Section 5: Visualizing Intelligence (Insights in Motion)

## 41.	The control room wants a visual trace of voltage stability over time.

```python
plt.figure(figsize=(12, 4))
sns.lineplot(x='timestamp', y='voltage_level', data=df_cleaned)
plt.title('Voltage Stability Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Voltage Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 42.	Pricing strategists need to evaluate how dynamic pricing flows throughout the day.

```python
df_cleaned['hour'] = df_cleaned['timestamp'].dt.hour
 
plt.figure(figsize=(10, 4))
sns.lineplot(x='hour', y='price_signal', data=df_cleaned, estimator='mean')
plt.title('Dynamic Pricing Flow by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Price Signal')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 43.	Forecast engineers request plots showing alignment between prediction and reality.

```python
plt.figure(figsize=(12, 4))
sns.lineplot(x='timestamp', y='historical_avg_demand', data=df_cleaned, label='Predicted')
sns.lineplot(x='timestamp', y='simulated_actual_demand', data=df_cleaned, label='Actual')
plt.title('Predicted vs Actual Demand')
plt.xlabel('Timestamp')
plt.ylabel('Demand')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 44.	Renewable integration needs a visual that shows how total demand is satisfied from each source.

```python
plt.pie(df_cleaned.groupby('energy_source')['simulated_actual_demand'].sum(), autopct='%1.1f%%', labels=['solar', 'wind', 'hydro'])
```

## 45.	Risk analysts are investigating whether usage clusters too aggressively â€” suggest visual proof.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_cleaned, x='historical_avg_demand', y='simulated_actual_demand', hue='Region')
plt.title('Usage Clustering by Region')
plt.xlabel('Historical Avg Demand')
plt.ylabel('Actual Demand')
plt.tight_layout()
plt.show()
```

## 46.	The planning team wants an overview of relationships between all numeric metrics.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_cleaned.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Numeric Metrics')
plt.tight_layout()
plt.show()
```

## 47.	Environmental teams suspect humidity affects voltage zones â€” they need validation.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_cleaned, x='operational_bands', y='humidity')
plt.title('Humidity Across Voltage Stability Zones')
plt.xlabel('Voltage Stability Band')
plt.ylabel('Humidity')
plt.tight_layout()
plt.show()
```

## 48.	Grid health reports require multi-angle snapshots of several metrics over the same time window.

```python
grid_long = df_cleaned.melt(id_vars='timestamp', value_vars=['voltage_level', 'grid_frequency', 'historical_avg_demand'], var_name='metric', value_name='value')
 
plt.figure(figsize=(14, 5))
sns.lineplot(data=grid_long, x='timestamp', y='value', hue='metric')
plt.title('Grid Health Metrics Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 49.	CFOs ask whether high prices correlate with demand spikes â€” visualize this.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_cleaned, x='price_signal', y='historical_avg_demand', hue='slot')
plt.title('Demand vs Dynamic Pricing by Time Slot')
plt.xlabel('Price Signal')
plt.ylabel('Actual Demand')
plt.tight_layout()
plt.show()
```

## 50.	Forecasting teams want to understand where their models fail â€” show a deviation pattern.

```python
plt.figure(figsize=(12, 4))
sns.lineplot(x='timestamp', y='deviation', data=df_cleaned, color='red')
plt.title('Forecast Deviation Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Absolute Deviation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
