```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Example of Data Visulization

```python
df = pd.DataFrame({
    'A': [1,2,3,4],
    'B': [5,6,7,8]
})

df.plot(kind='bar')
plt.show()
```

```python
df = pd.DataFrame({
    'Month': ['Jan','Feb','Mar','Apr'],
    'Energy_Usage': [280,300,250,360]
})

df.set_index('Month').plot(kind='bar', title='Monthly Energy Usage')
plt.ylabel('kWh')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x,y,marker='^')
plt.title('Simple Line Chart')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.figure(figsize=(8, 5)) # Set figure size (width, height in inches)
plt.plot(x, y, label='Sales', color='blue', linestyle='--', marker='o', linewidth=2, markersize=8)

plt.title("Monthly Sales", fontsize=16, color='darkblue') # Title with font size and color
plt.xlabel("Month", fontsize=12)
plt.ylabel("Revenue ($K)", fontsize=12)

plt.xticks(rotation=45) # Rotate x-axis labels
plt.yticks(fontsize=10)

plt.grid(True, linestyle=':', linewidth=0.7, color='gray') # Add gridlines
plt.legend(loc='upper left') # Add legend
plt.tight_layout() # Adjust layout to prevent overlap
plt.savefig("sales_plot.png", dpi=300) # Save plot
plt.show()
```
