# Example of Data Viz

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'Energy_Usage': [300, 280, 350, 400]
})
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.figure(figsize=(8, 5))  # Set figure size (width, height in inches)
plt.plot(x, y, label='Sales', 
         color='blue',
         linestyle='--',
         marker='o',
         linewidth=2, markersize=8)

plt.title("Monthly Sales", fontsize=16, color='darkblue')  # Title with font size and color
plt.xlabel("Month", fontsize=12)
plt.ylabel("Revenue ($K)", fontsize=12)

plt.xticks(rotation=45)  # Rotate x-axis labels
plt.yticks(fontsize=10)

plt.grid(True, linestyle=':', linewidth=0.7, color='gray')  # Add gridlines
plt.legend(loc='upper left')  # Add legend
plt.tight_layout()  # Adjust layout to prevent overlap
plt.savefig("sales_plot.png", dpi=300)  # Save plot
plt.show()
```

# Bar plot

```python
categories = ['A', 'B', 'C']
values = [100, 150, 90]

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.title("Category Performance", fontsize=14)
plt.xlabel("Category")
plt.ylabel("Score")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2, 1, figsize=(6, 6))  # 2 rows, 1 column

axs[0].plot(x, y1, color='blue')
axs[0].set_title("Sine Wave")

axs[1].plot(x, y2, color='green')
axs[1].set_title("Cosine Wave")

fig.suptitle("Trigonometric Functions")
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6), sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axs[i, j].plot(x, np.sin(x + i + j))
        axs[i, j].set_title(f"Plot ({i},{j})")

fig.suptitle("2x2 Grid of Sine Waves")
plt.tight_layout()
plt.show()
```

# 1️⃣ Line Plot — Trend Over Time

```python
#Use Case: Show how energy usage changes month by month.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
north_usage = [300, 320, 310, 330, 340, 360]

plt.plot(months, north_usage, marker='o', color='blue')
plt.title("North Zone Energy Usage Over Time")
plt.xlabel("Month")
plt.ylabel("Energy (kWh)")
plt.grid(True)
plt.show()
```

# 2️⃣ Bar Chart — Compare Categories

```python
#Use Case: Compare energy usage across neighborhoods in June.
zones = ['North', 'South', 'East', 'West']
june_usage = [360, 400, 380, 370]

plt.bar(zones, june_usage, color='skyblue')
plt.title("Energy Usage by Zone (June)")
plt.ylabel("Energy (kWh)")
plt.show()
```

# 3️⃣ Histogram — Distribution of Daily Usage

```python
import numpy as np
daily_usage = np.random.normal(loc=50, scale=10, size=100)
```

```python
# Use Case: Show how daily usage varies in the South zone.
plt.hist(daily_usage, bins=10, color='orange', edgecolor='black')
plt.title("Distribution of Daily Energy Usage (South Zone)")
plt.xlabel("Daily Usage (kWh)")
plt.ylabel("Frequency")
plt.show()
```

# 4️⃣ Box Plot — Summary Statistics

```python
data = [
    [300, 320, 310, 330, 340, 360],  # North
    [400, 420, 410, 430, 440, 460],  # South
    [380, 390, 370, 400, 410, 420],  # East
    [370, 360, 350, 380, 390, 400]   # West
]
```

```python
plt.boxplot(data, labels=zones)
plt.title("Energy Usage Variability by Zone")
plt.ylabel("Energy (kWh)")
plt.grid(True)
plt.show()
```

# 5️⃣ Scatter Plot — Correlation Between Temp & Usage


```python
#Use Case: Explore relationship between temperature and energy usage.
temperature = [20, 22, 24, 26, 28, 30]
usage = [300, 320, 340, 360, 390, 420]

plt.scatter(temperature, usage, color='green')
plt.title("Temperature vs. Energy Usage")
plt.xlabel("Temperature (°C)")
plt.ylabel("Energy (kWh)")
plt.grid(True)
plt.show()
```

# 6️⃣ Pie Chart — Proportion of Total Usage

```python
#Use Case: Show each zone’s contribution to total energy usage.
total_usage = [2100, 2400, 2300, 2200]

plt.pie(total_usage, labels=zones, autopct='%1.1f%%', startangle=90)
plt.title("Energy Usage Share by Zone")
plt.axis('equal')  # Equal aspect ratio ensures a perfect circle
plt.show()
```

# 7️⃣ Area Plot — Cumulative Usage Over Time

```python
# Use Case: Show how total energy usage accumulates over months.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
north = [300, 320, 310, 330, 340, 360]
south = [400, 420, 410, 430, 440, 460]

plt.stackplot(months, north, south, labels=['North', 'South'], colors=['blue', 'red'])
plt.title("Cumulative Energy Usage Over Time")
plt.ylabel("Energy (kWh)")
plt.legend(loc='upper left')
plt.show()
```

# ⚡ Project Title: “Smart City Energy Dashboard”

```python
import pandas as pd
import numpy as np

np.random.seed(42)
# Create date range
dates = pd.date_range(start='2023-01-01', end='2023-06-30', freq='D')
# Zones
zones = ['North', 'South', 'East', 'West']
# Generate synthetic daily usage data (in kWh)
data = {
    'Date': np.tile(dates, len(zones)),
    'Zone': np.repeat(zones, len(dates)),
    'Usage_kWh': np.random.normal(loc=350, scale=50, size=len(dates) * len(zones)).round(2)
}
df = pd.DataFrame(data)
df['Usage_kWh'] = df['Usage_kWh'].clip(lower=200)  # No negative usage
```

```python
df
```

```python
df.info()
```

# 1️⃣ Line Plot — Time Series Trend by Zone

```python
import matplotlib.pyplot as plt 
```

```python
df_pivot = df.pivot(index='Date', columns='Zone', values='Usage_kWh')
# default plot function have kind is line plot 
df_pivot.plot(figsize=(12, 6), title='Daily Energy Usage by Zone')
plt.ylabel('Usage (kWh)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 2️⃣ Bar Chart — Average Usage by Zone

```python
avg_usage = df.groupby('Zone')['Usage_kWh'].mean()
avg_usage.plot(kind='bar', color='skyblue', title='Average Energy Usage by Zone')
plt.ylabel('Average Usage (kWh)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# 3️⃣ Box Plot — Variability in Usage

```python
df.boxplot(column='Usage_kWh', by='Zone', figsize=(8, 6))
plt.title('Usage Variability by Zone')
plt.suptitle('')
plt.ylabel('Usage (kWh)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 4️⃣ Histogram — Distribution of Usage

```python
df[df['Zone'] == 'South']['Usage_kWh'].plot(kind='hist', bins=20,
                                            color='orange',
                                            edgecolor='black')
plt.title('Distribution of Daily Usage (South Zone)')
plt.xlabel('Usage (kWh)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# create subplot for all 4 zones distribution of usage
x = [['North', 'South'], ['East', 'West']]

fig, axs = plt.subplots(2, 2, figsize=(10, 8), sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        df[df['Zone'] == x[i][j]]['Usage_kWh'].plot(kind='hist', bins=20,
                                                          color='orange',
                                                          edgecolor='black',
                                                          ax=axs[i][j])
        axs[i, j].set_title(f"Distribution of Daily Usage ({x[i][j]} Zone)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 5️⃣ Pie Chart — Total Usage Share

```python
total_usage = df.groupby('Zone')['Usage_kWh'].sum()
total_usage.plot(kind='pie', autopct='%1.1f%%',
                 startangle=90,
                 title='Total Usage Share by Zone')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
