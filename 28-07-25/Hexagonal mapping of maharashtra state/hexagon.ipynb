```python
# importing all the necessary libraries
import h3
import pandas as pd
import folium
import branca.colormap as cm
```

```python
# Setting coordinates and resolution
lat = 19.7515
long = 75.7139

polygon = [
    (22.012428, 74.408566), #bottom-left
    (20.094454, 72.754605), #bottom-right
    (15.756535, 73.703200), #top-right
    (18.764097, 80.124459), #top-left
    (19.510900, 80.866309),
    (21.300342, 80.635241),
    (22.012428, 74.408566)
]


resolution = 4
```

```python
# converting coordinates to h3
poly = h3.LatLngPoly(polygon)
h3_indexes = h3.polygon_to_cells(poly, resolution)
print(h3_indexes)
print(poly)
```

```python
# Convert H3 to coordinates
lat_lng = [h3.cell_to_latlng(h) for h in h3_indexes]
print(f"Latitude / Longitude: {lat_lng}")
```

```python
# Convert h3 to Geometry Boundary
cell_boundary = []
for h3_index in h3_indexes:
    boundary = h3.cell_to_boundary(h3_index)
    cell_boundary.append(boundary)
    
# boundary = h3.cell_to_boundary(large_hex)
print(cell_boundary)
```

```python
# Mapping the boundary
m = folium.Map(location = [lat, long], zoom_start=6)

# Add the H3 cell to the map
folium.Polygon(locations=cell_boundary, color="blue", weight=5).add_to(m)

# Save the map
m
```

```python
df = pd.read_csv('Longitude_and_Latitude_of_Important_locations_in_Maharashtra.csv')
df
```

```python
def dms_to_decimal(dms):
    parts = dms.replace('°', ' ').replace("'", '').split()
    deg = int(parts[0])
    min = int(parts[1])
    direction = parts[2]
    
    decimal = deg + min / 60
    return round(decimal, 6)
```

```python
df['Latitude_decimal'] = df['Latitude'].apply(dms_to_decimal)
df['Longitude_decimal'] = df['Longitude'].apply(dms_to_decimal)
df
```

```python
def compute_h3_and_boundaries(row, resolution=4):
    lat = row['Latitude_decimal']
    lng = row['Longitude_decimal']
    h3_index = h3.latlng_to_cell(lat, lng, resolution)
    boundaries = h3.cell_to_boundary(h3_index)
    return pd.Series([h3_index, boundaries])
```

```python
df[['H3_Index', 'Boundaries']] = df.apply(compute_h3_and_boundaries, axis=1)
```
